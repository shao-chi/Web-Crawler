```python
import requests
from bs4 import BeautifulSoup 
```

```python
url="https://rate.bot.com.tw/xrt/quote/2018-01/USD"

r = requests.get(url)
print(r.text) #印出HTML
```

```python
soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
sel = soup.select("td.text-center") #取HTML標中的 <td class="text-center"></td> 中的<a>標籤存入sel
buy_in = soup.select("td.rate-content-cash")
# sell_out = soup.select("td.rate-content-cash")
print(len(sel), len(buy_in))
for i in range(0,len(sel),2):
    print(sel[i].text, buy_in[i].text, buy_in[i+1].text)#, sell_out[i].text) 
```

```python
data = []
for i in range(1,10):
    url="https://rate.bot.com.tw/xrt/quote/2018-0{}/USD".format(i)
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
    sel = soup.select("td.text-center") #取HTML標中的 <td class="text-center"></td> 中的<a>標籤存入sel
    buy_in = soup.select("td.rate-content-cash")
    # sell_out = soup.select("td.rate-content-cash")
    print(len(sel)/2, len(buy_in)/2)
    for i in range(0,len(sel),2):
        data.append({'date': sel[i].text, 'buy_in': buy_in[i].text, 'sell_out': buy_in[i+1].text})
        print(sel[i].text, buy_in[i].text, buy_in[i+1].text)#, sell_out[i].text) 
        
for i in range(10,13):
    url="https://rate.bot.com.tw/xrt/quote/2018-{}/USD".format(i)
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
    sel = soup.select("td.text-center") #取HTML標中的 <td class="text-center"></td> 中的<a>標籤存入sel
    buy_in = soup.select("td.rate-content-cash")
    # sell_out = soup.select("td.rate-content-cash")
    print(len(sel)/2, len(buy_in)/2)
    for i in range(0,len(sel),2):
        data.append({'date': sel[i].text, 'buy_in':buy_in[i].text, 'sell_out': buy_in[i+1].text})
        print(sel[i].text, buy_in[i].text, buy_in[i+1].text)#, sell_out[i].text) 
        
url="https://rate.bot.com.tw/xrt/quote/2019-01/USD"
print(url)
r = requests.get(url)
soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
sel = soup.select("td.text-center") #取HTML標中的 <td class="text-center"></td> 中的<a>標籤存入sel
buy_in = soup.select("td.rate-content-cash")
# sell_out = soup.select("td.rate-content-cash")
print(len(sel)/2, len(buy_in)/2)
for i in range(0,len(sel),2):
    data.append({'date': sel[i].text, 'buy_in':buy_in[i].text, 'sell_out': buy_in[i+1].text})
    print(sel[i].text, buy_in[i].text, buy_in[i+1].text)#, sell_out[i].text) 
```

```python
import pandas as pd
```

```python
data_df = pd.DataFrame(data)
data_df.to_csv('usd201801~.csv', encoding='utf-8')
print(data_df)
```
